```python
import pandas as pd
import numpy as np
```

```python
t = pd.read_csv('Data.csv', delimiter=',')
del t['Отметка времени']
names = t.columns
```

```python
events = dict.fromkeys(t['В каком мероприятии вы принимали участие?'].value_counts().to_frame().index, dict())
```

```python
events
```

```python
options = ['Квест "Cercopidae"', 'Викторина к дню народного единства 13.11 на ИСиРе']
out = []
for option in options:
    out.append(t[t[names[3]].isin([option])])
```

```python
for i in out:
    for j in i[names[4]].value_counts().index:
        event = i[names[3]].value_counts().index[0]
        print(i[names[3]].value_counts().index[0])
        people = i[i[names[4]]==j][names[0]].tolist()
        print(j)
        print(people)
        print()
        print(events)
        print('-'*100)
        events[event][j] = people
```

```python
events
```

```python
other_values={
    'Квест "Cercopidae"': ['Квест', 'Квест "Cercopidae"', "8.12.2021"],
    'Викторина к дню народного единства 13.11 на ИСиРе': ["Викторина", "Вместе - мы едины", "13.11.2021"],
}
```

```python
if len(events) != len(other_values):
    print("Внимание!\nВнесите в переменную \"other_values\" данные в формате \[формат, название, дата\]")
```

```python
s = "\n-----------------------------------\n\n"
for i in events:
    arr = other_values[i]
    s += f"Формат: {arr[0]}\nНазвание: {arr[1]}\nДата проведения: {arr[2]}\n\n"
    
    position_on_event = list(events[i].keys())
    position_on_event.append('Участник')
    position_on_event = sorted(position_on_event, reverse=True)
    if position_on_event[0] == "Участник":
        position_on_event = position_on_event[1:] + [position_on_event[0]]
    
    for j in position_on_event:
        if j in events[i]:
            count = 1
            ss = ''
            names = sorted(events[i][j])
            for name in names:
                ss += f"{count}. {name};\n"
                count += 1
            ss = ss[:-1]+'.\n\n'
            
            s += f"{j}:\n{ss}"
        
    
    s += "\n-----------------------------------\n\n"
```

```python
print(s)
```

```python
with open('Квест и викторина.txt', 'w') as f:
    f.write(s)
```
